```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

y = np.linspace(0,8*np.pi)
x = np.sin(y)

fig, ax = plt.subplots()
ax.axis([0,4*np.pi,-1.5,1.8])
l, = ax.plot([],[])

def animate(i):
    l.set_data(y[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(y))

from IPython.display import HTML
HTML(ani.to_jshtml())
```
